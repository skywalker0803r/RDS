```python
feed.columns = ['feed' + str(i) for i in feed.columns]
op.columns = ['op' + str(i) for i in op.columns]
ProductDSAR.columns = ['ProductDSAR' + str(i) for i in ProductDSAR.columns]
ProductDSLGO.columns = ['ProductDSLGO' + str(i) for i in ProductDSLGO.columns]
```

```python
merged_df = pd.concat([feed,op,ProductDSAR,ProductDSLGO],axis=1)
merged_df
```

```python
merged_df.dropna(inplace=True)
merged_df.shape
```
